Ahora vamos a explorar los datos faltantes

```python
#Filtrar filas con algun dato faltante
filas_con_na = data[data.isna().any(axis=1)]
print(f"Total filas con al menos un nulo: {filas_con_na.shape[0]}")
filas_con_na.head()
```

Al analizar el dataset por filas nos dimos cuenta que 703 filas contienen al menos un valor nulo, lo cual es significativo teniendo en cuenta el total de observaciones. Esto implica que eliminar todas las filas con datos faltantes no sera una estrategia correcta , dado que reducira el tamao de los datos de entrenamiento y se podra incrementar la varianza del modelo o introducir sesgo si los valores faltantes no son completamente aleatorios. Por otro lado, habiamos mencionado antes que el target tiene valores nulos , asi que estas filas las vamos a eliminar, dado que no es apropiado imputar la variable dependiente en un modelo supervisado. 

```python
filas_sin_na = data.dropna()

print("Media con nulos:")
print(filas_con_na.describe())

print("Media sin nulos:")
print(filas_sin_na.describe())
```

La comparacin entre filas con y sin valores faltantes nos deja ver que las estadsticas descriptivas no presentan diferencias sustanciales, esto puede sugerir que los datos faltantes no parecen alterar significativamente la distribucin de las variables. Por esto, se prefiere imputar datos en las variables independientes en vez de eliminarlos.
